```python
import nbfinder
from loader import g
from loader import T
from loader import __
```

## groupCount()
__Get types of vertices__

```python
x1 = g.V().groupCount().by(T.label).next()
x1
```

Equivalent to

```python
x2 = g.V().label().groupCount().next()
assert x1 == x2
```

## group()
Also equivalent to

```python
x3 = g.V().group().by(T.label).by(__.count()).next()
assert x2 == x3
```

__Count of airports by country__

```python
g.V().hasLabel('airport').groupCount().by('country').next()
```

In select countries

```python
g.V().hasLabel('airport').groupCount().by('country').select('FR','GR','BE').next()
```

__Count of outgoing vertices for each continent code__

```python
g.V().hasLabel('continent').group().by('code').by(__.out().count()).next()
```

__Group all descriptions by their label__

```python
g.V().group().by(T.label).by('desc').next()
```

__Runways grouped by code__

```python
g.V().hasLabel('airport').limit(5).group().by('code').by('runways').next()
```
